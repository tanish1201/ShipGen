```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("E:/sem5/apl302product/computational_project/reduced_data.csv")
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Select features (all PCA columns in this case)
X = df.iloc[:, :]

# Apply K-Means clustering (choosing 3 clusters as an example; adjust as needed)
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

# Add the cluster labels to the data
df['Cluster'] = kmeans.labels_

# Reduce dimensions for visualization (2D)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the data in 2D with clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis', s=50, marker='o')
plt.title("2D Visualization of Clusters with K-Means")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label="Cluster Label")
plt.show()
```

```python
from sklearn.manifold import TSNE
```

```python
# Apply t-SNE for dimensionality reduction to 2D
tsne = TSNE(n_components=2, random_state=0)
X_tsne = tsne.fit_transform(X)

# Plot the data in 2D with clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=kmeans.labels_, cmap='viridis', s=50, marker='o')
plt.title("2D Visualization of Clusters with t-SNE")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.colorbar(label="Cluster Label")
plt.show()
```

```python
# Select features
X = df.iloc[:, :]

# Determine the optimal number of clusters using the elbow method
inertia = []
cluster_range = range(1, 11)  # Testing from 1 to 10 clusters

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)  # Inertia is the WCSS metric

# Plot the elbow graph
plt.figure(figsize=(8, 6))
plt.plot(cluster_range, inertia, marker='o', linestyle='-')
plt.title("Elbow Method for Optimal Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia (WCSS)")
plt.xticks(cluster_range)
plt.show()
```

```python
# Choose the optimal cluster number based on the elbow graph (let's say it's 3 here)
optimal_clusters = 4
kmeans = KMeans(n_clusters=optimal_clusters, random_state=0)
kmeans.fit(X)

# Add the optimal cluster labels to the data
df['Cluster'] = kmeans.labels_

# Reduce dimensions for visualization (2D)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the data in 2D with clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis', s=50, marker='o')
plt.title(f"2D Visualization of Clusters with K-Means (k={optimal_clusters})")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label="Cluster Label")
plt.show()
```
